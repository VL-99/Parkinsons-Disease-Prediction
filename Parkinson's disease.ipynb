# Parkinson's Disease Predictor#

###We use different machine learning algorithms and models to check if a person is going to be effected by the Parkinson's disease or not### 

#Importing Libraries#

```python
import pandas
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pandas.read_csv('parkinsons.data')
```

#Data Exploration

###Statistics of people suffering

```python
plt.figure(figsize=(10,6))
plt.hist(df.status)
plt.xlabel('Status', fontsize=16)
plt.ylabel('Number of ppl',fontsize=16)
plt.title('Histogram of number of people suffering with Parkinson syndrome', fontsize=16)
```

```python
df.drop(labels=['name'], axis=1, inplace=True)
```

#Splitting the data#

```python
X=df.drop(labels=['status'], axis=1)
y=df['status']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, random_state=40)
```

#Machine Learning#

###Logistic Regression###

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score
regressor=LogisticRegression(random_state=40)
regressor.fit(X_train, y_train)
```

```python
y_train_pred=regressor.predict(X_train)
print(confusion_matrix(y_train, y_train_pred))
accuracy_score(y_train, y_train_pred)
```

```python
y_test_pred=regressor.predict(X_test)
print(confusion_matrix(y_test, y_test_pred))
accuracy_score(y_test,y_test_pred)
```

##Decision Tree classifier##

```python
from sklearn.tree import DecisionTreeClassifier
DT=DecisionTreeClassifier()
DT.fit(X,y)
```

```python
y_train_pred_dt=DT.predict(X_train)
print(confusion_matrix(y_train, y_train_pred_dt))
accuracy_score(y_train,y_train_pred_dt)
```

```python
y_test_pred_dt=DT.predict(X_test)
print(confusion_matrix(y_test, y_test_pred_dt))
accuracy_score(y_test,y_test_pred_dt)
```

##Naive Bayes method##

```python
from sklearn.naive_bayes import GaussianNB
NB=GaussianNB()
NB.fit(X_train, y_train)
```

```python
y_train_pred_nb=NB.predict(X_train)
print(confusion_matrix(y_train, y_train_pred_nb))
accuracy_score(y_train,y_train_pred_nb)
```

```python
y_test_pred_nb=NB.predict(X_test)
print(confusion_matrix(y_test, y_test_pred_nb))
accuracy_score(y_test,y_test_pred_nb)
```

##Support Vector Machines##

```python
from sklearn.svm import SVC
SVM=SVC(kernel='linear')
SVM.fit(X_train,y_train)
```

```python
y_train_pred_svm=SVM.predict(X_train)
print(confusion_matrix(y_train, y_train_pred_svm))
accuracy_score(y_train,y_train_pred_svm)
```

```python
y_test_pred_svm=SVM.predict(X_test)
print(confusion_matrix(y_test, y_test_pred_svm))
accuracy_score(y_test,y_test_pred_svm)
```

##Random Forest##

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier().fit(X_train,y_train)
```

```python
y_train_pred_rf=classifier.predict(X_train)
print(confusion_matrix(y_train, y_train_pred_rf))
accuracy_score(y_train,y_train_pred_rf)
```

```python
y_test_pred_rf=classifier.predict(X_test)
print(confusion_matrix(y_test, y_test_pred_rf))
accuracy_score(y_test,y_test_pred_rf)
```

```python
df_compare=pandas.DataFrame(data=[y_test_pred_rf, y_test])
df_compare.T
```

##K-Nearest Neighbours##

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier()
classifier.fit(X_train,y_train)
```

```python
y_train_pred_knn=classifier.predict(X_train)
print(confusion_matrix(y_train, y_train_pred_knn))
accuracy_score(y_train,y_train_pred_knn)
```

```python
y_test_pred_knn=classifier.predict(X_test)
print(confusion_matrix(y_test, y_test_pred_knn))
accuracy_score(y_test,y_test_pred_knn)
```

#Out of all the models implemented, random forest and decision tree algorithms worked well on this data giving a high accuracy.
